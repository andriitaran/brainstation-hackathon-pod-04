```python
from sklearn.model_selection import train_test_split
X_train_lifestyle, X_test_lifestyle, y_train_lifestyle, y_test_lifestyle = train_test_split(X_lifestyle, y_lifestyle, test_size=0.3, random_state=1)

X_train_entertainment, X_test_entertainment, y_train_entertainment, y_test_entertainment = train_test_split(X_entertainment, y_entertainment, test_size=0.3, random_state=1)

X_train_business, X_test_business, y_train_business, y_test_business = train_test_split(X_business, y_business, test_size=0.3, random_state=1)

X_train_socmed, X_test_socmed, y_train_socmed, y_test_socmed = train_test_split(X_socmed, y_socmed, test_size=0.3, random_state=1)

X_train_tech, X_test_tech, y_train_tech, y_test_tech = train_test_split(X_tech, y_tech, test_size=0.3, random_state=1)

X_train_world, X_test_world, y_train_world, y_test_world = train_test_split(X_world, y_world, test_size=0.3, random_state=1)
```

```python
from sklearn.preprocessing import StandardScaler
scaler_lifestyle = StandardScaler()
scaler_lifestyle.fit(X_train_lifestyle)
X_train_lifestyle = scaler_lifestyle.transform(X_train_lifestyle)
X_test_lifestyle = scaler_lifestyle.transform(X_test_lifestyle)

scaler_entertainment = StandardScaler()
scaler_entertainment.fit(X_train_entertainment)
X_train_entertainment = scaler_entertainment.transform(X_train_entertainment)
X_test_entertainment = scaler_entertainment.transform(X_test_entertainment)

scaler_business = StandardScaler()
scaler_business.fit(X_train_business)
X_train_business = scaler_business.transform(X_train_business)
X_test_business = scaler_business.transform(X_test_business)

scaler_socmed = StandardScaler()
scaler_socmed.fit(X_train_socmed)
X_train_socmed = scaler_socmed.transform(X_train_socmed)
X_test_socmed = scaler_socmed.transform(X_test_socmed)

scaler_tech = StandardScaler()
scaler_tech.fit(X_train_tech)
X_train_tech = scaler_tech.transform(X_train_tech)
X_test_tech = scaler_tech.transform(X_test_tech)

scaler_world = StandardScaler()
scaler_world.fit(X_train_world)
X_train_world = scaler_world.transform(X_train_world)
X_test_world = scaler_world.transform(X_test_world)
```

Function taking input from website and generating "good" or "bad" rating

```python
# Title input:
def title_count(title):
    count = 0
    nlp = English()
    doc = nlp(title)
    for i in doc:
        count += 1
    return count
```

```python
# Body input:
def body_counts(body):
    count = 0
    nunique = 0
    n_non_sw = 0
    n_non_sw_unique = 0
    unique = []
    non_sw = []
    non_sw_unique = []
    nlp = English()
    doc = nlp(body)
    for i in doc:
        count += 1
        if str(i) not in unique:
            unique.append(str(i))
        if i.is_stop == False:
            non_sw.append(i)
        if (i.is_stop == False) & (str(i) not in non_sw_unique):
            non_sw_unique.append(str(i))
    nunique = len(unique)
    n_non_sw = len(non_sw)
    n_non_sw_unique = len(non_sw_unique)
    return count, nunique, n_non_sw, n_non_sw_unique
```

```python
def prediction_info(title,
                    body,
                    num_hrefs,
                    num_self_hrefs,
                    num_imgs,
                    num_videos):
    n_tokens_title = title_count(title)
    n_tokens_content, n_unique_tokens, n_non_stop_words, n_non_stop_unique_tokens = body_counts(body)
    df_predict = pd.DataFrame({' n_tokens_title':n_tokens_title,
                               ' n_tokens_content':n_tokens_content, 
                               ' n_unique_tokens':n_unique_tokens, 
                               ' n_non_stop_words':n_non_stop_words, 
                               ' n_non_stop_unique_tokens':n_non_stop_unique_tokens,
                               ' num_hrefs':num_hrefs, 
                               ' num_self_hrefs':num_self_hrefs,
                               ' num_imgs':num_imgs, 
                               ' num_videos':num_videos}, index=[0])
    return df_predict
```

```python
def userinput(user_input, category):
    if category == 'lifestyle':
        output = pd.concat([user_input, dfmean_lifestyle], axis=1)
    elif category == 'entertainment':
        output = pd.concat([user_input, dfmean_entertainment], axis=1)
    elif category == 'business':
        output = pd.concat([user_input, dfmean_business], axis=1)
    elif category == 'socmed':
        output = pd.concat([user_input, dfmean_socmed], axis=1)
    elif category == 'tech':
        output = pd.concat([user_input, dfmean_tech],axis=1)
    elif category == 'world':
        output = pd.concat([user_input, dfmean_world], axis=1)
    return output
```
